```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wavf
```

```python
with open("./pfeiffen.pkl", "rb") as myfile:
    data = np.array(pickle.load(myfile))

'''with open("./voice.pkl", "rb") as myfile:
    data = np.array(pickle.load(myfile))'''

'''with open("./sprache.pkl", "rb") as myfile:
    data = np.array(pickle.load(myfile))'''

print(np.min(data))
print(np.max(data))
print(np.mean(data))

plt.figure()
plt.plot(data)
plt.show()
```

```python
scaled_data = ((data-np.mean(data))/np.max(np.abs(data-np.mean(data)))*32767).astype(np.int16)

plt.figure()
plt.plot(scaled_data)
plt.show()
```

```python
fs = int(np.round(len(scaled_data)/5))
out_f = 'out.wav'

wavf.write(out_f, fs, scaled_data)
```
